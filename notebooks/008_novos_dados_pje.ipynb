# Teste de extração de informação de processo

```python
path = '../data/raw/exporta_pje_20250520'
path_processo = os.path.join(path, 'processo')

# Lista e todos os arquivos .txt
processos = [f for f in os.listdir(path_processo) if f.endswith('.xml')]

# Pega o primeiro arquivo (não está ordenado como na pasta)
primeiro_processo = processos[0]
caminho_processo = os.path.join(path_processo, primeiro_processo)

def xml_file_to_string(file_path):
    tree = ET.parse(file_path)
    root = tree.getroot()

    xml_string = ET.tostring(root, encoding='utf-8').decode('utf-8')

    return xml_string

result_test = xml_file_to_string(caminho_processo)

# Lê o conteudo do aviso
#with open(test_path, 'r', encoding='utf-8', errors='replace') as f:
#    content = f.read()

#result = extrair_informacao_do_xml_processo(caminho_processo)
```

# Teste de extração de informações de aviso

```python
path = '../data/raw/exporta_pje_20250520'
path_aviso = os.path.join(path, 'aviso')

# Lista e todos os arquivos .txt
avisos = [f for f in os.listdir(path_aviso) if f.endswith('.txt')]

# Pega o primeiro arquivo (não está ordenado como na pasta)
primeiro_aviso = avisos[0]
caminho_aviso = os.path.join(path_aviso, primeiro_aviso)

caminho_aviso
```

# Extraindo os processos

```python
import pandas as pd
from tqdm import tqdm

list_df_processo = []

for processo in tqdm(processos):
    path = os.path.join(path_processo, processo)

    try:
        xml_text = xml_file_to_string(path)
        list_df_processo.append(extrair_informacao_do_xml_processo(xml_text))
    
    except Exception as e:
        print(f'Erro {e}')
        list_df_processo.append(pd.DataFrame())
```

```python
df_processo = pd.concat(list_df_processo)
df_processo.to_csv(os.path.join('../data/raw/exporta_pje_20250520', 'novos_processos.csv'))
```

# Extraindo Avisos

```python
list_df_avisos = []

for aviso in tqdm(avisos):
    path = os.path.join(path_aviso, aviso)

    try:
        with open(path, 'r', encoding='latin1') as f:
            content = f.read()
            
        list_df_avisos.append(extrair_informacao_do_xml_avisos(content))
    
    except Exception as e:
        print(f'Erro {e}')
        list_df_avisos.append(pd.DataFrame())
```

```python
df_avisos = pd.concat(list_df_avisos)
df_avisos.to_csv(os.path.join('../data/raw/exporta_pje_20250520', 'novos_avisos.csv'))
```
